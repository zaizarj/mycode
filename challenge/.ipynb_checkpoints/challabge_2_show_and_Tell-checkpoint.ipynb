```python
import matplotlib.pyplot as plt
import pandas as pd
mydf = pd.read_csv("cereal.csv", sep=",")
print(mydf)
mydf.plot(x='name', rot=75)
plt.title("Most Unhealthy Cereals")
plt.xlabel("Cereal Brands")
plt.ylabel("Amount")
```

```python
mydf = mydf[["name", "sugars","carbo"]]
print(mydf)
mydf.plot(x='name', rot=75)
plt.title("Most Unhealthy Cereals")
plt.xlabel("Cereal Brands")
plt.ylabel("Amount")
```
