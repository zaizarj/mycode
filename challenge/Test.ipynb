```python
# Use subplots() to specify titles and labels

       # generate the numbers 1 through 9 and convert that to a list
       # [1,2,3,4,5,6,7,8,9]
days = range(1, 9)
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]

fig, ax = plt.subplots()
ax.plot(days, celsius_values)
ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')

plt.show()
```

```python
# Use the same axes for multiple lines of data

days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]

fig, ax = plt.subplots()

ax.set(xlabel='Day',
       ylabel='Temperature in Celsius',
       title='Temperature Graph')


ax.plot(days, celsius_min,
        days, celsius_min, "oy",
        days, celsius_max,
        days, celsius_max, "or")

plt.show()
```

```python
# Make a bar plot of visitors over time (years)

        # this is a list comprehension; the contents of the list are being generated inside the list itself
        # "for every number between 2010 and 2021, add that number to this list"
years = [str(year) for year in range(2010, 2021)]
visitors = (1241, 50927, 162242, 222093, 
            665004, 2071987, 2460407, 3799215, 
            5399000, 5474016, 6003672)


plt.bar(years, visitors, color="green")

plt.xlabel("Years")
plt.ylabel("Values")
plt.title("Bar Chart Example")

plt.plot()
plt.show()
```

```python
# Create randomly generated dummy data that follows a normal distribution
# https://en.wikipedia.org/wiki/Normal_distribution

gaussian_numbers = np.random.normal(size=10000)
gaussian_numbers

                          
plt.hist(gaussian_numbers, bins=20) # bins are the number of "columns" that make up the distribution curve
plt.title("Gaussian Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a stacked plot for indices 1-9

idxes = [ 1,  2,  3,  4,  5,  6,  7,  8,  9]
y1  = [23, 42, 33, 43,  8, 44, 43, 18, 21]
y2  = [9, 31, 25, 14, 17, 17, 42, 22, 28]
y3  = [18, 29, 19, 22, 18, 16, 13, 32, 1]


plt.stackplot(idxes, y1, y2, y3)
plt.title('Stack Plot Example')

plt.show()
```

```python
# Pie charts are generally considered poor ways to communicate data visually, so proceed with caution!!

labels = 'Python', 'C', 'Java', 'C++', 'C#'
sizes = [14.26, 13.06, 11.19, 8.66, 5.92]
explode = (0.1, 0, 0, 0, 0)  # how far a pie slice "sticks out"

fig1, ax1 = plt.subplots()
ax1.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%',  # this autopct argument takes float values up-to tenth place after decimal
        shadow=True,        # shadow draws a shadow beneath the pie
        startangle=0)       # startangle rotates the start of the pie chart by angle degrees counterclockwise from the x-axis                              
                                              
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('TIOBE Index for March 2022')
plt.show()
```
